## STEP 1 — Install and Load spaCy Model

```python

!pip install spacy --quiet
!python -m spacy download en_core_web_sm

import spacy

nlp = spacy.load("en_core_web_sm")
```

## STEP 2 — Sentence Dataset

```python

sentences = [
    "The cat sits on the mat.",
    "She is reading a book.",
    "The boy who is playing is my brother.",
    "He bought a car and drove it home.",
    "Why are you late today?",
    "The ball was kicked by the player.",
    "AI is transforming the world.",
    "I love programming in Python.",
    "The teacher explained the lesson clearly.",
    "They went to the market and bought fruits.",
    "This is a complex sentence because it has multiple clauses.",
    "The sun rises in the east.",
    "She quickly finished her homework.",
    "Can you solve this problem?",
    "The project was completed successfully."
]
```

## STEP 3 — Dependency Parsing

```python

import pandas as pd

results = []

for sent in sentences:
    doc = nlp(sent)
    for token in doc:
        results.append([
            sent,
            token.text,
            token.lemma_,
            token.pos_,
            token.dep_,
            token.head.text
        ])

df = pd.DataFrame(results, columns=[
    "Sentence", "Token", "Lemma", "POS", "Dependency", "Head"
])

df.head(20)
```

## STEP 4 — Sentence Structure Analysis

```python

for sent in sentences:
    doc = nlp(sent)

    subject = [token.text for token in doc if token.dep_ == "nsubj"]
    obj = [token.text for token in doc if token.dep_ == "dobj"]
    root = [token.text for token in doc if token.dep_ == "ROOT"]

    print("Sentence:", sent)
    print("Subject:", subject)
    print("Root:", root)
    print("Object:", obj)
    print("-"*50)
```

## STEP 5 — Dependency Tree Visualization

```python

from spacy import displacy

for sent in sentences[:5]:
    doc = nlp(sent)
    displacy.render(doc, style="dep", jupyter=True)
```

## STEP 6 — Comparative Analysis


Simple sentences have fewer dependencies and a clear subject-verb-object structure.
Complex sentences contain multiple clauses and more dependency links.
Modifiers such as adjectives and adverbs increase the depth of the dependency tree.
Passive voice changes subject-object relationships.


## STEP 7 — Conclusion


spaCy dependency parsing helps understand grammatical structure.
It identifies relationships like subject, object, and modifiers.
Visualization improves interpretability of sentence structure.
